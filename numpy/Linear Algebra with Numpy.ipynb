## Doing Linear Algebra while Learning Numpy

```python
import numpy as np
```

```python
A = np.arange(9).reshape(3,3)
B = np.arange(start = 8, stop = -1, step = -1).reshape(3,3)
```

```python
A
```

```python
B
```

```python
# Hadamand Product
C = np.multiply(A, B)
```

```python
C
```

```python
# Matrix Multiplication
D = A @ B
```

```python
K = np.array([[3, 4], [1, 1]])
```

```python
K
```

```python
# Inverse of K
K_inv = np.linalg.inv(K)
```

```python
K_inv
```

```python
K @ K_inv
```

```python
#Identity Matrix
I = np.eye(3)
```

```python
I
```

```python
v = np.array([1, 5])
```

```python
w = K @ v
```

```python
w.shape
```

```python
# why is this valid??
# Is v a column vector or a row vector?
z = v @ K
```

```python
z
```

```python
w @ z
```

```python
# This seems to be more explicit
# and better?
w.reshape(2,1) @ z.reshape(1,2)
```
